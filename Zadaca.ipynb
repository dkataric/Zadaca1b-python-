## **Vremenske promjene od početka mjerenja**

```python
import pandas as pd
globalne = pd.read_csv('GlobalTemperatures.csv')
globalne.columns
```

```python
globalne.head()
```

```python
len(globalne.index) / 12
```

```python
prosjek_po_godinama = {}

LandAverage = globalne['LandAverageTemperature'].tolist()

for i in range(266):
    prosjek = 0
    for j in range(12):
        prosjek = prosjek + LandAverage[i + j]
    prosjek_po_godinama[i+1750] = prosjek/12
```

```python
LandAv, ax1 = plt.subplots()

ax1.plot(prosjek_po_godinama.keys(), prosjek_po_godinama.values())

ax1.set(xlabel='godina mjerenja', ylabel='temperatura u celzijevima stupnjevima',
       title='Promjena temperature planeta kroz povijest')
ax1.grid()
plt.ylim(5,11)
```

Vidimo kako zemljina površina nema neočekivan/značajan trend rasta ili pada.

Od 1850.godine imamo podatke za sve stupce pa ćemo pogledati maksimalnu i minimalnu temperaturu po godinama.

```python
prosjek_po_najvecoj = {}
prosjek_po_najmanjoj = {}

LandMax = globalne.dropna()['LandMaxTemperature'].tolist()
LandMin = globalne.dropna()['LandMinTemperature'].tolist()
```

```python
for i in range(len(LandMax)//12):
    prosjek_min = 0
    prosjek_max = 0
    for j in range(12):
        prosjek_min = prosjek_min + LandMin[i + j]
        prosjek_max = prosjek_max + LandMax[i + j]
    prosjek_po_najvecoj[i+1850] = prosjek_max/12
    prosjek_po_najmanjoj[i+1850] = prosjek_min/12
    
LandMaxMin, ax2 = plt.subplots()
ax2.grid()
ax2.plot(prosjek_po_najvecoj.keys(), prosjek_po_najvecoj.values())
ax2.plot(prosjek_po_najmanjoj.keys(), prosjek_po_najmanjoj.values())
```

Niti kod maksimalnih i minimalnih temperatura nije zabilježen značajniji trend rasta/pada od početka mjerenja temperature.

Temperatura u Hrvtaskoj očito ima trend rasta. Stavit ćemo na isti graf i prosjek za prvi (najhladniji) mjesec te za sedmi (najtopliji) i usporediti da vidimo povećava li se temperatura više zbog toplijih ljeta ili toplije zime.

```python
prosjek_7 = {}
prosjek_1 = {}
suma1 = 0
suma7 = 0
pom = 0

for i in range (125543,128771,12):
    #ne gledati godine u kojoj nedostaje podatak
    if (not math.isnan(hrv.iloc[i,1]) and  not math.isnan(hrv.iloc[i+6,1])):
        pom = pom + 1
        suma1 = suma1 + hrv.iloc[i,1]
        suma7 = suma7 + hrv.iloc[i+6,1]
        if (pom == 10):
            prosjek_1[int(hrv.iloc[i,0][:4])] = suma1/10
            prosjek_7[int(hrv.iloc[i,0][:4])] = suma7/10
            suma1 = 0
            suma7 = 0
            pom = 0
        
LandHrv_1_7, ax5 = plt.subplots()
ax5.plot(prosjek_hrv_novi.keys(), prosjek_hrv_novi.values(), label = 'srednji prosjek')
ax5.plot(prosjek_1.keys(), prosjek_1.values(), label = 'prosjek za 1.mj')
ax5.plot(prosjek_7.keys(), prosjek_7.values(), label = 'prosjek za 7.mj')
ax5.set(xlabel='godina mjerenja', ylabel='temperatura u celzijevima stupnjevima')
ax5.legend(loc=2)
ax5.grid()
```

Očito su povećanjem temperature podjednako doprinijela i toplija ljeta, ali i toplije zime zbog podjednakog rasta u temperaturi što je bilo i očekivano.

## **CO2 u atmosferi**

Analizirajući podtake sasvim je jasno kako zemlja ima periode toplijih i hladnijih klima na globalnom nivou (gledajući cijelu povijest planeta). Kad uzmemo pojedini period opet vidimo toplija/hladnija razdoblja, te tako možemo doći skroz do pojedinog dana kada se izmjenjuju toplija(dan) i hladnija(noć) razdoblja.
No, ono što je zanimljivo kad se gleda globalno zagrijavanje je to da su sve vremenske promjene bile prirodne za planetu zemlju i odvijale su se bez utjecaja ljudi kroz jako dugi vremenski period (milijuni godina). Ono što se danas događa je promjena klime zbog umjetnih utjecaja koji nemaju veze sa prirodnom zemljinom klimom nego zbog povećane razine CO2 u atmosferi i to u nekoliko desetaka godina.
